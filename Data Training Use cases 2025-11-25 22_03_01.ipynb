```python
# Github code integration
#I. Python Operators Usecases
#Use Case 1: Internet Data Usage Calculator


totalmonthlydata:int = int(input("Enter the total monthly data in GB: "))
usedmonthlydata:float = float(input("Enter the used monthly data in GB: "))
remainingdata = round(totalmonthlydata - usedmonthlydata,2)
usagepercentage:float = round((usedmonthlydata / totalmonthlydata) * 100,2)
print(f"Remaining data : ", float(remainingdata))
print(f"Used Percentage : ",usagepercentage)
if usagepercentage >= 80:
    print(f"Your used Percentage is",usagepercentage, "Warning: High usage, consider upgrading your plan.")
```

```python
#Use Case 2: Shopping Discount Calculation
originalprice = input("Enter the original price")
discountpercentage = input("Enter the discountpercentage")
discountprice = float(originalprice)*(float(discountpercentage)/100)
finalprice = float(originalprice) - discountprice
print(f"Original price is", originalprice)
print(f"Discounted price is", discountprice)
print(f"Final price is", finalprice)
```
